```python
import numpy as np
import matplotlib.pyplot as plt
import TopHat as th
from PIL import Image
import RoCurvetest as roc
```

```python
ruta = '/home/denisse/Downloads/drive/DRIVE/training/images/39_training.tif'
img = Image.open(ruta).convert("L")
img_array = np.array(img)
```

```python
kernel1 = th.crea_kernel(11)
TopHat = th.BlackHat(img_array, kernel1)
```

```python
gt_ruta = '/home/denisse/Downloads/drive/DRIVE/training/1st_manual/39_manual1.gif'
gtim = Image.open(gt_ruta).convert("L")
gt_array = np.array(gtim)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

ax[0].imshow(TopHat, cmap='gray')
ax[0].set_title("Result")
ax[0].axis('off')

ax[1].imshow(gt_array, cmap='gray')
ax[1].set_title("Ground Truth")
ax[1].axis('off')

plt.show()
```

```python
sensitivity = np.zeros(255)
specificity = np.zeros(255)
for i in range(0, 255):
    sen, spe = roc.tptn(i, TopHat, gt_array)
    sensitivity[i] = sen
    specificity[i] = spe
```

```python
roc.curva_roc(sensitivity, specificity)
```
