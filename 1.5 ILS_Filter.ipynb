```python
import numpy as np
import matplotlib.pyplot as plt
import TopHat as th
from PIL import Image 
import ILS  

import Metrics as roc 
```

```python
ruta = 'drive/DRIVE/training/images/39_training.tif'
img = Image.open(ruta).convert("L")
img_array = np.array(img)
```

```python
kernel_inicial = th.crea_kernel(25)
TopHat = th.BlackHat(img_array, kernel_inicial)
```

```python
gt_ruta = 'drive/DRIVE/training/1st_manual/39_manual1.gif'
gtim = Image.open(gt_ruta).convert("L")
gt_array = np.array(gtim)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

ax[0].imshow(TopHat, cmap='gray')
ax[0].set_title("Result")
ax[0].axis('off')

ax[1].imshow(gt_array, cmap='gray')
ax[1].set_title("Ground Truth")
ax[1].axis('off')

plt.show()
```

```python
roc.ROC_value(TopHat, gt_array)
```

```python
best_kernel, best_Az = ILS.iterated_local_search(kernel_inicial, img_array, gt_array, max_iter=10)

print("\nValor Az asociado:", best_Az)

print("Mejor kernel encontrado:\n")
plt.imshow(best_kernel, cmap='gray')
```
